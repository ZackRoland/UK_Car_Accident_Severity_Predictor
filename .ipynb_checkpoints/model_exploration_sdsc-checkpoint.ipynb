```python
bins = [0, 30, 50, 70]
labels = ['Low', 'Medium', 'High']
df['Speed_limit_binned'] = pd.cut(
    df['Speed_limit'],
    bins=bins,
    labels=labels,
    include_lowest=True
)
df['Speed_limit_binned'] = df['Speed_limit_binned'].cat.codes.replace(-1, np.nan)

# Fill missing categorical values
df['Junction_Control'] = df['Junction_Control'].fillna('No junction')
df['Carriageway_Hazards'] = df['Carriageway_Hazards'].fillna('No Hazards')
df['Special_Conditions_at_Site'] = df['Special_Conditions_at_Site'].fillna('No special Conditions')

# Drop missing in crossings
df = df.dropna(subset=['Pedestrian_Crossing-Human_Control'])
df = df.dropna(subset=['Pedestrian_Crossing-Physical_Facilities'])

# Fix typo in Light_Conditions
df['Light_Conditions'] = df['Light_Conditions'].replace(
    {'Darkeness: No street lighting': 'Darkness: No street lighting'}
)

# Drop missing in key columns
df = df.dropna(subset=['Time', 'Weather_Conditions', 'Road_Surface_Conditions', 'Did_Police_Officer_Attend_Scene_of_Accident'])

# --- Diagnostic output ---
print("\n[Speed_limit_binned] unique values and null count:")
print(df['Speed_limit_binned'].unique(), "Nulls:", df['Speed_limit_binned'].isna().sum())

print("\n[Junction_Control] unique values:")
print(df['Junction_Control'].unique())

print("\n[Carriageway_Hazards] unique values:")
print(df['Carriageway_Hazards'].unique())

print("\n[Special_Conditions_at_Site] unique values:")
print(df['Special_Conditions_at_Site'].unique())

print("\n[Pedestrian_Crossing-Human_Control] uniques and nulls:")
print(df['Pedestrian_Crossing-Human_Control'].unique(), "Nulls:", df['Pedestrian_Crossing-Human_Control'].isna().sum())

print("\n[Pedestrian_Crossing-Physical_Facilities] uniques and nulls:")
print(df['Pedestrian_Crossing-Physical_Facilities'].unique(), "Nulls:", df['Pedestrian_Crossing-Physical_Facilities'].isna().sum())

print("\n[Light_Conditions] unique values:")
print(df['Light_Conditions'].unique())

print("\nFinal DataFrame shape:", df.shape)
print("Remaining nulls per column:")
print(df.isna().sum())
```

```python
accidentdf1 = pd.read_csv(Path(root) / "accidents_2005_to_2007.csv" ,low_memory = False)
accidentdf2 = pd.read_csv(Path(root) / "accidents_2009_to_2011.csv",low_memory = False)
accidentdf3 = pd.read_csv(Path(root) / "accidents_2012_to_2014.csv",low_memory = False)

df_concat = pd.concat([accidentdf1, accidentdf2, accidentdf3])
```

```python
# --- 1. DATA PREPROCESSING ---
df = df_concat.drop(columns=[
    'Accident_Index','Location_Easting_OSGR','Location_Northing_OSGR','Date',
    'LSOA_of_Accident_Location','Local_Authority_(District)','Local_Authority_(Highway)',
    '1st_Road_Number','Longitude','Latitude','1st_Road_Class','Road_Type','Police_Force',
    'Junction_Detail','2nd_Road_Class','2nd_Road_Number'
])

df['Junction_Control'] = df['Junction_Control'].fillna('No junction')
df['Light_Conditions'] = df['Light_Conditions'].replace(
    {'Darkeness: No street lighting': 'Darkness: No street lighting'}
)
df['Special_Conditions_at_Site'] = df['Special_Conditions_at_Site'].fillna('None')
df['Carriageway_Hazards'] = df['Carriageway_Hazards'].fillna('None')
df = df.dropna()
scaler = sk.preprocessing.StandardScaler()
df[['Number_of_Vehicles', 'Number_of_Casualties']] = scaler.fit_transform(
    df[['Number_of_Vehicles', 'Number_of_Casualties']]
)
df = pd.get_dummies(df, columns=[
    'Day_of_Week', 'Time', 'Speed_limit', 'Junction_Control', 'Pedestrian_Crossing-Human_Control',
    'Road_Surface_Conditions', 'Pedestrian_Crossing-Physical_Facilities',
    'Light_Conditions', 'Weather_Conditions', 'Special_Conditions_at_Site', 
    'Carriageway_Hazards', 'Urban_or_Rural_Area',
    'Did_Police_Officer_Attend_Scene_of_Accident', 'Year'
])
x_train, x_test, y_train, y_test = sk.model_selection.train_test_split(
    df.drop(columns=['Accident_Severity']),
    df['Accident_Severity'],
    test_size=0.2,
    random_state=42
)
print("Data preprocessing complete.")
```

```python
start_nb = time.time()
print("Training Naive Bayes...")
naive_bayes = sk.naive_bayes.GaussianNB()
naive_bayes.fit(x_train, y_train)

print("\nEvaluating TRAINING performance...")
y_hat_nb_train = naive_bayes.predict(x_train)
report_nb_train = sk.metrics.classification_report(y_train, y_hat_nb_train, output_dict=True)
report_nb_train = pd.DataFrame(report_nb_train).transpose()
print("Training Set Performance:")
print(report_nb_train)

print("\nPredicting TEST Naive Bayes output...")
y_hat_nb = []
for row in tqdm(np.array_split(x_test, 100), desc="NaiveBayes Predict Progress"):
    y_hat_nb.extend(naive_bayes.predict(row))
y_hat_nb = np.array(y_hat_nb)

report_nb_test = sk.metrics.classification_report(y_test, y_hat_nb, output_dict=True)
report_nb_test = pd.DataFrame(report_nb_test).transpose()
print("Test Set Performance:")
print(report_nb_test)

end_nb = time.time()
print(f"\nNaive Bayes completed in {end_nb - start_nb:.2f} seconds.")

# Compare Training vs Test accuracy
train_acc = float(report_nb_train.loc["accuracy"].values[0])
test_acc = float(report_nb_test.loc["accuracy"].values[0])

print(f"\nTraining accuracy: {train_acc:.4f}")
print(f"Test accuracy:    {test_acc:.4f}")
```

```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score, classification_report
```

```python
smoothing_values = [1e-9, 1e-8, 1e-7, 1e-6]
train_accuracies = []
test_accuracies = []
train_reports = []
test_reports = []

start_nb = time.time()
print("Training and Evaluating Naive Bayes with varying var_smoothing:")

for v in smoothing_values:
    print(f"\nvar_smoothing={v:.0e}:")
    nb = sk.naive_bayes.GaussianNB(var_smoothing=v)
    nb.fit(x_train, y_train)

    # Training performance
    y_train_pred = nb.predict(x_train)
    train_acc = accuracy_score(y_train, y_train_pred)
    train_accuracies.append(train_acc)
    train_report = classification_report(y_train, y_train_pred, output_dict=True)
    train_reports.append(pd.DataFrame(train_report).transpose())
    print(f"Training Accuracy: {train_acc:.4f}")

    # Test performance (progress bar over chunks)
    y_test_pred = []
    for chunk in tqdm(np.array_split(x_test, 100), desc=f"Predicting TEST output (vs={v:.0e})"):
        y_test_pred.extend(nb.predict(chunk))
    y_test_pred = np.array(y_test_pred)

    test_acc = accuracy_score(y_test, y_test_pred)
    test_accuracies.append(test_acc)
    test_report = classification_report(y_test, y_test_pred, output_dict=True)
    test_reports.append(pd.DataFrame(test_report).transpose())
    print(f"Test Accuracy:    {test_acc:.4f}")

end_nb = time.time()
print(f"\nNaive Bayes grid training & evaluation completed in {end_nb - start_nb:.2f} seconds.")

# Plot train vs test accuracy for visual fitting graph
# Plot train vs test accuracy for visual fitting graph
plt.figure(figsize=(8,5))
plt.plot(smoothing_values, train_accuracies, label='Train accuracy', marker='o')
plt.plot(smoothing_values, test_accuracies, label='Test accuracy', marker='o')
plt.xscale('log')
plt.xlabel('var_smoothing')
plt.ylabel('Accuracy')
plt.legend()
plt.title('NB Classification: Train vs Test Accuracy')

# ---- Save the figure before showing ----
plt.savefig("nb_fitting_curve.png", dpi=300)  # saves to current directory
plt.show()


# Best model selection and fitting analysis
best_idx = np.argmax(test_accuracies)
print(f"\nBest var_smoothing: {smoothing_values[best_idx]:.0e}")
print(f"Best Test Accuracy: {test_accuracies[best_idx]:.4f}")
print("Classification Report for Best Model (Test Set):")
print(test_reports[best_idx])
```
